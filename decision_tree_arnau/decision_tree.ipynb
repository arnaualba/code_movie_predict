```python
import os
import pandas as pd
```

```python
data = pd.read_csv('../../merge_imdb_id_data.csv')
data
```

```python
data = data.drop(['Unnamed: 0', 'original_title_y', 'revenue', 'tagline'], axis = 1)
data
```

```python
def date_to_num (date):
    date = pd.to_datetime(date, format='%Y-%m-%d')
    new_year_day = pd.Timestamp(year=date.year, month=1, day=1)
    return (date - new_year_day).days + 1
    
if os.path.isfile('date_to_num.csv'):
    data = pd.read_csv('date_to_num.csv')
else:
    for date in data['release_date']:
        date = str(date)
        if '-' in date:
            data = data.replace( date, date_to_num(date) )
    data.to_csv('date_to_num.csv')
    
data = data.dropna(axis = 0)
data
```

```python
from sklearn import model_selection as ms
from sklearn import preprocessing, neighbors, svm
import numpy as np

x = np.array(data[['release_date', 'startYear', 'runtimeMinutes', 'budget']])
y = np.array(data['averageRating'])
for i in range(len(y)):
    if y[i] < 7:
        y[i] = 0
    else:
        y[i] = 1
            

x_train, x_test, y_train, y_test = ms.train_test_split(x, y, test_size=0.2)

clf = svm.SVC()
clf.fit(x_train, y_train)
confidence = clf.score(x_test, y_test)
print('confidence = ' + str(confidence))
print(clf)
```
