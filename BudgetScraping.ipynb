```python
from requests import get
from requests.exceptions import RequestException
from contextlib import closing
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = get("https://www.the-numbers.com/movies/year/2019")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')
Data = pd.read_html(str(table))
```

```python
if len(Data)==1:
    for d in Data: data=d
```

```python
data
```

```python
data=data.drop(['ReleaseDate'], axis=1)
data=data.drop(['Trailer'], axis=1)
data=data.drop(['Genre'], axis=1)
```

```python
for col in data.columns:            #removes any row with missing data
    data=data[data[col]!='$0']
data=data[~data.ReleaseType.str.contains("2019", na=False)]
data.dropna(how='all')
```

```python
s = "https://www.the-numbers.com/movies/year/"

D = pd.DataFrame()

for y in range(1980,2020):
    
    print(y)
    string=s+str(y)
    
    res = get(string)
    soup = BeautifulSoup(res.content,'lxml')
    table = soup.find_all('table')
    Data = pd.read_html(str(table))
    
    if len(Data)==1:
        for d in Data: df=d
    
    df=df.drop(['ReleaseDate'], axis=1)
    df=df.drop(['Trailer'], axis=1)
    df=df.drop(['Genre'], axis=1)

    for col in df.columns:            #removes any row with missing data
        df=df[df[col]!='$0']
    df=df[~df.ReleaseType.str.contains(str(y), na=False)]
        
    D=pd.concat([D,df])
```
