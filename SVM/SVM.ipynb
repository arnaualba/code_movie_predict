```python
import os
import pandas as pd
```

```python
data = pd.read_csv('../../merge_imdb_id_data.csv')
data
```

```python
data = data.drop(['Unnamed: 0', 'original_title_y', 'revenue', 'tagline'], axis = 1)
data
```

```python
def date_to_num (date):
    date = pd.to_datetime(date, format='%Y-%m-%d')
    new_year_day = pd.Timestamp(year=date.year, month=1, day=1)
    return (date - new_year_day).days + 1
    
if os.path.isfile('date_to_num.csv'):
    data = pd.read_csv('date_to_num.csv')
else:
    for date in data['release_date']:
        date = str(date)
        if '-' in date:
            data = data.replace( date, date_to_num(date) )
    data.to_csv('date_to_num.csv')
    
data = data.dropna(axis = 0)
data
```

```python
from sklearn import model_selection as ms
from sklearn import preprocessing, neighbors, svm
import numpy as np

x = np.array(data[['release_date', 'startYear']])
y = np.array(data['averageRating'])
for i in range(len(y)):
    if y[i] < 7:
        y[i] = 0
    else:
        y[i] = 1
            

x_train, x_test, y_train, y_test = ms.train_test_split(x, y, test_size=0.2)

clf = svm.SVC()
clf.fit(x_train, y_train)
confidence = clf.score(x_test, y_test)
print('confidence = ' + str(confidence))
print(clf)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

fig, ax = plt.subplots()
for i in range(len(y_train)):
    if i % 100 == 0:
        if y_train[i]:
            ax.scatter(x_train[i][0], x_train[i][1], color = 'r')
        else: 
            ax.scatter(x_train[i][0], x_train[i][1], color = 'b')
plt.xlabel('day of the year')
plt.ylabel('year')
legend_elements = [Line2D([0], [0], marker = 'o', label = 'Score < 7', markerfacecolor='b', markersize=10),
                   Line2D([0], [0], marker = 'o', label = 'Score > 7', markerfacecolor='r', markersize=10)]
ax.legend(handles=legend_elements, loc='upper left')
plt.show()
```

```python
# Trying to predict some stuff

# Darknknight : 9.0,
# Avengers inf war : 8.5,
# Fast and F 8 : 6.7,
# Star Wars VII : 8.0
example_measures = np.array([ [date_to_num('2008-07-18'), 2008], [date_to_num('2018-04-27'), 2018],\
                             [date_to_num('2017-04-14'), 2008], [date_to_num('2015-12-18'), 2008] ]) 

prediction = clf.predict(example_measures)
print(prediction)
```
